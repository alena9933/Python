```python
df = pd.read_csv('productstar_df.csv')
df
```

```python
df['shipped_at'] = pd.to_datetime(df['shipped_at'])
df['completed_at'] = pd.to_datetime(df['completed_at'])
df1 = df[df['order_state'] == 'complete'].copy()
```

```python
df1['date'] = df1['completed_at'].dt.round('D')
df1['week'] = df1['date'] - pd.to_timedelta(df1['date'].dt.dayofweek, unit='day') 
df1['year'] = df1['date'].dt.year
# df1['weeknum'] = df1['date'].dt.week
# df1['days_en_route'] = (df1['shipped_at'] - df1['completed_at']).dt.days
```

```python
# Если будет несколько столбцов, то нарисует несколько графиков
df_t = (df1[(df1['date']>='2017-01-01') & (df1['date']<'2020-01-01')]
        .groupby(['store_id'])
        .sum()[['item_count']]
        .unstack())
# df_t.columns = df_t.columns.get_level_values(1)
df_t.plot(figsize=(20,10)) # у plot есть кучка параметров
```
